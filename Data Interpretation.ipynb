```python
# importing libraries
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing data.
df = pd.read_excel('Data_Interpretation_Section_1.xlsx')
pd.set_option('display.max_rows', 56)
#dropping NaN values
df = df.fillna(0)
df
```

```python
#Scaling variables in million.
df['1990 enroll'] = df['1990 enroll'].div(1000000)

df['1990 gdp'] = df['1990 gdp'].div(1000000000)

df['1991 enroll'] = df['1991 enroll'].div(1000000)

df['1991 gdp'] = df['1991 gdp'].div(1000000000)

df['1992 enroll'] = df['1992 enroll'].div(1000000)

df['1992 gdp'] = df['1992 gdp'].div(1000000000)

df['1993 enroll'] = df['1993 enroll'].div(1000000)

df['1993 gdp'] = df['1993 gdp'].div(1000000000)

df['1994 enroll'] = df['1994 enroll'].div(1000000)

df['1994 gdp'] = df['1994 gdp'].div(1000000000)

df['1995 gdp'] = df['1995 gdp'].div(1000000000)

df['1995 enroll'] = df['1995 enroll'].div(1000000)

df['1997 enroll'] = df['1997 enroll'].div(1000000)

df['1996 enroll'] = df['1996 enroll'].div(1000000)

df['1996 gdp'] = df['1996 gdp'].div(1000000000)

df['1997 gdp'] = df['1997 gdp'].div(1000000000)

df['1998 enroll'] = df['1998 enroll'].div(1000000)

df['1998 gdp'] = df['1998 gdp'].div(1000000000)

df['1999 enroll'] = df['1999 enroll'].div(1000000)

df['1999 gdp'] = df['1999 gdp'].div(1000000000)

df['2000 enroll'] = df['2000 enroll'].div(1000000)

df['2000 gdp'] = df['2000 gdp'].div(1000000000)
df
```

```python
df.plot(x='countryname', y=['1990 enroll', '1991 enroll', '1992 enroll', '1993 enroll', '1994 enroll',
                           '1995 enroll', '1996 enroll', '1997 enroll', '1998 enroll', '1999 enroll',
                           '2000 enroll'], kind = 'bar', figsize=(20,10), linewidth=5, fontsize=20)
```
